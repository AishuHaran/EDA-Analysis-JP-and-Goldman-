### **Goldman Trading Volume Patterns**

```python
df_goldman.dtypes
```

```python
# Function to convert 'Volume' to a numeric type
def convert_volume(value):
    if isinstance(value, str):  # Check if the value is a string
        if 'M' in value:
            return float(value.replace('M', '')) * 1_000_000  # Convert millions
        elif 'K' in value:
            return float(value.replace('K', '')) * 1_000  # Convert thousands
    return float(value)  # If already numeric, convert to float

# Apply conversion to the 'Volume' column
df_goldman['Volume'] = df_goldman['Volume'].apply(convert_volume)

# Now, calculate correlation
correlation = df_goldman[['Change %', 'Volume']].corr().iloc[0, 1]
print(f"Correlation between price change and volume: {correlation:.2f}")
```

```python
# Calculate the correlation
correlation = df_goldman[['Change %', 'Volume']].corr().iloc[0, 1]
print(f"Correlation between price change and volume: {correlation:.2f}")

# Plot the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_goldman, x='Volume', y='Change %', alpha=0.5)
plt.title('Correlation Between Trading Volume and Price Change %: Goldman')
plt.xlabel('Volume')
plt.ylabel('Price Change %')
plt.show()
```

### **JPMorgan Trading Volume Patterns**

```python
# Apply conversion to the 'Volume' column
df_jp['Volume'] = df_jp['Volume'].apply(convert_volume)

# Now, calculate correlation
correlation = df_jp[['Change %', 'Volume']].corr().iloc[0, 1]
print(f"Correlation between price change and volume: {correlation:.2f}")
```

```python
# Plot the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_jp, x='Volume', y='Change %', alpha=0.5)
plt.title('Correlation Between Trading Volume and Price Change %: JPMorgan')
plt.xlabel('Volume')
plt.ylabel('Price Change %')
plt.show()
```

### **Calculate Pearson Correlation & P-Value**

```python
import scipy.stats as stats

# Ensure numeric data (convert 'Volume' if needed)
df_jp['Volume'] = pd.to_numeric(df_jp['Volume'], errors='coerce')

# Calculate Pearson correlation and p-value
corr_coeff, p_value = stats.pearsonr(df_jp['Change %'], df_jp['Volume'])

print(f"Pearson Correlation for JPMorgan: {corr_coeff:.3f}")
print(f"P-Value for JPMorgan: {p_value:.5f}")
```

```python
# Ensure numeric data (convert 'Volume' if needed)
df_goldman['Volume'] = pd.to_numeric(df_goldman['Volume'], errors='coerce')

# Calculate Pearson correlation and p-value
corr_coeff, p_value = stats.pearsonr(df_goldman['Change %'], df_goldman['Volume'])

print(f"Pearson Correlation for Goldman: {corr_coeff:.3f}")
print(f"P-Value for Goldman: {p_value:.5f}")
```

**Final Conclusion for JPMorgan**
- There is a statistically significant relationship between trading volume and price change.
- However, the correlation is so small (0.026) that it has little to no practical impact—trading volume alone is not a strong predictor of price movement for JPMorgan.
- Other factors (e.g., macroeconomic news, interest rates, investor sentiment) might have a stronger influence on price changes.

**Final Conclusion for Goldman Sachs**
- There is a statistically significant but very weak negative correlation between trading volume and price change.
- This means that although the correlation is statistically significant, the relationship is too weak to be of practical use in predicting price changes based on trading volume.

### **Goldman Comparison of Open, High, Low, and Close Prices**

#### **Change % Analysis:**  
   - What are the most frequent percentage changes?  
   - Are there extreme changes that indicate major events affecting stock prices?

```python
# Calculate daily percentage change
df_jp['Pct_Change'] = df_jp['Close'].pct_change() * 100

# Analyze most frequent percentage changes
most_frequent_changes = df_jp['Pct_Change'].round(2).value_counts().head(10)
```

```python
# Identify extreme changes (e.g., beyond ±15%)
extreme_changes = df_jp[(df_jp['Pct_Change'] > 15) | (df_jp['Pct_Change'] < -15)]

# Visualization: Histogram of percentage changes
plt.figure(figsize=(10,5))
plt.hist(df_jp['Pct_Change'].dropna(), bins=50, edgecolor='k', alpha=0.7)
plt.axvline(df_jp['Pct_Change'].mean(), color='r', linestyle='dashed', linewidth=2, label="Mean Change")
plt.xlabel('Percentage Change')
plt.ylabel('Frequency')
plt.title('Distribution of Stock Price Percentage Changes; JPMorgan')
plt.legend()
plt.show()
```

```python
# Print Results
print("Most Frequent Percentage Changes:\n", most_frequent_changes)
print("\nExtreme Changes:\n", extreme_changes)
```

**Inference**
- Frequent Small Price Changes: The most frequent percentage changes are small, with 0.00% appearing 41 times, indicating that the stock often shows little to no movement day-to-day.

#### **Volatility Comparison**



```python
# Calculate daily returns for both companies
df_goldman['Daily_Return'] = df_goldman['Close'].pct_change()
df_jp['Daily_Return'] = df_jp['Close'].pct_change()

# Calculate volatility (standard deviation of returns)
JP_volatility = df_jp['Daily_Return'].std()
Goldman_volatility = df_goldman['Daily_Return'].std()

print(f"Goldman Sachs Volatility: {Goldman_volatility}")
print(f"JPMorgan Chase Volatility: {JP_volatility}")
```

```python
# Calculate rolling volatility (e.g., 30-day window)
df_goldman['Rolling_Volatility'] = df_goldman['Daily_Return'].rolling(window=30).std()
df_jp['Rolling_Volatility'] = df_jp['Daily_Return'].rolling(window=30).std()

# Plot rolling volatility
plt.figure(figsize=(14, 7))
plt.plot(df_goldman.index, df_goldman['Rolling_Volatility'], label='Goldman Sachs Volatility', color='blue')
plt.plot(df_jp.index, df_jp['Rolling_Volatility'], label='JPMorgan Chase Volatility', color='red')

plt.title('30-Day Rolling Volatility: Goldman Sachs vs. JPMorgan Chase')
plt.xlabel('Date')
plt.ylabel('Rolling Volatility')
plt.legend()
plt.grid(True)
plt.show()
```

##### **Inference**
- Goldman Sachs is more volatile, reacting sharply to market events like the COVID-19 crash (2020) and possibly the early 2000s dot-com bust or financial regulation changes (2003).
- JPMorgan, despite fluctuations, shows more controlled volatility, likely due to its diversified banking model (retail + investment).
- The average volatility values (GS: 0.0227, JPM: 0.0236) are close, but Goldman experiences sharper, short-term spikes, whereas JPM maintains a steadier trajectory over time.

#### **Comparison of Trading Volume**


```python
plt.figure(figsize=(14, 7))

# Plot trading volumes
plt.plot(df_goldman.index, df_goldman['Volume'], label='Goldman Sachs Volume', color='blue')
plt.plot(df_jp.index, df_jp['Volume'], label='JPMorgan Chase Volume', color='red')

plt.title('Comparison of Trading Volume: Goldman Sachs vs. JPMorgan Chase')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate daily percentage change in trading volume
df_goldman['Volume_pct_change'] = df_goldman['Volume'].pct_change() * 100  # percentage change in volume
df_jp['Volume_pct_change'] = df_jp['Volume'].pct_change() * 100

# Define a threshold for "big jumps" (e.g., mean + 2 standard deviations)
threshold_goldman = df_goldman['Volume_pct_change'].mean() + 2 * df_goldman['Volume_pct_change'].std()
threshold_jpmorgan = df_jp['Volume_pct_change'].mean() + 2 * df_jp['Volume_pct_change'].std()

# Filter for big jumps
df_goldman_big_jumps = df_goldman[df_goldman['Volume_pct_change'] > threshold_goldman]
df_jp_big_jumps = df_jp[df_jp['Volume_pct_change'] > threshold_jpmorgan]
```

```python
# Sort and get the top 10 biggest jumps for Goldman Sachs
top_10_goldman = df_goldman_big_jumps.sort_values(by='Volume_pct_change', ascending=False).head(10)
print("Top 10 Biggest Jumps for Goldman Sachs:")
print(top_10_goldman[['Volume', 'Volume_pct_change']])

# Sort and get the top 10 biggest jumps for JPMorgan Chase
top_10_jp = df_jp_big_jumps.sort_values(by='Volume_pct_change', ascending=False).head(10)
print("\nTop 10 Biggest Jumps for JPMorgan Chase:")
print(top_10_jp[['Volume', 'Volume_pct_change']])
```

Significant Volatility Periods:

#### **Quarterly Closing Price for Goldman Sachs and JPMorgan Chase**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Resample data to quarterly frequency and calculate the average closing price for each quarter
df_goldman_quarterly = df_goldman['Close'].resample('Q').mean()
df_jp_quarterly = df_jp['Close'].resample('Q').mean()

# Plot the quarterly closing prices for both Goldman Sachs and JPMorgan Chase
plt.figure(figsize=(12, 6))
plt.plot(df_goldman_quarterly, label='Goldman Sachs Quarterly Average', color='blue')
plt.plot(df_jp_quarterly, label='JPMorgan Chase Quarterly Average', color='red')

# Adding labels and title
plt.title('Quarterly Closing Price for Goldman Sachs and JPMorgan Chase')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Group by year and find the quarter with the highest closing price for each year
goldman_max_quarter = df_goldman_quarterly.groupby(df_goldman_quarterly.index.year).idxmax()
jpmorgan_max_quarter = df_jp_quarterly.groupby(df_jp_quarterly.index.year).idxmax()

# Print out the quarterly averages for both companies
print("Goldman Sachs Top Quarterly Averages:\n", goldman_max_quarter)
print()
print("JPMorgan Chase Top Quarterly Averages:\n", jpmorgan_max_quarter)
```

##### **Inference**

- Both Goldman Sachs and JPMorgan Chase show recurring top quarterly averages at year-end (Q4), particularly in December. This suggests a strong seasonal effect, possibly driven by year-end financial reporting, market trends, or institutional rebalancing.  
- Goldman Sachs exhibits several peaks in Q1 (March), indicating strong performance early in the year, potentially due to earnings reports or market optimism at the start of new fiscal periods.  
- JPMorgan Chase has more variability in its top quarters, with peaks occurring in both Q1 (March) and Q3 (September), suggesting different cyclical factors impacting its stock performance.  
- The presence of 2020 and 2021 in both lists aligns with pandemic-driven market volatility and recovery periods.  
- Overall, while both stocks show strong Q4 trends, Goldman Sachs leans more towards Q1 strength, whereas JPMorgan Chase displays more scattered performance spikes across different quarters.
